```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Create a figure and axis
fig, ax = plt.subplots(figsize=(8, 6))
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.axis('off')

# Add boxes for each step
def add_box(text, xy, box_color='lightblue'):
    box = patches.FancyBboxPatch(xy, 4, 1, boxstyle="round,pad=0.1", fc=box_color, ec="black")
    ax.add_patch(box)
    ax.text(xy[0] + 2, xy[1] + 0.5, text, ha='center', va='center', fontsize=10)

# Add arrows between boxes
def add_arrow(start, end):
    ax.annotate('', xy=end, xytext=start, arrowprops=dict(arrowstyle='->', lw=2))

# Define positions
start_pos = (3, 9)
decision_pos = (3, 7)
yes_pos = (1, 5)
no_pos = (6, 5)
end_pos = (3, 3)

# Draw boxes
add_box("In√≠cio", start_pos)
add_box("Est√° chovendo?", decision_pos, box_color='lightyellow')
add_box("Pegar guarda-chuva", yes_pos)
add_box("N√£o precisa de guarda-chuva", no_pos)
add_box("Fim", end_pos)

# Draw arrows
add_arrow((5, 9), (5, 8))  # In√≠cio -> Decis√£o
add_arrow((3, 7), (2, 6))  # Decis√£o -> Sim
add_arrow((7, 7), (7, 6))  # Decis√£o -> N√£o
add_arrow((2, 5), (3, 4))  # Sim -> Fim
add_arrow((7, 5), (4, 4))  # N√£o -> Fim

# Add decision labels
ax.text(1.5, 6.5, "Sim", fontsize=10)
ax.text(6.5, 6.5, "N√£o", fontsize=10)

# Show the plot
plt.title("Fluxograma: Decis√£o de usar guarda-chuva", fontsize=12)
```
